# Continuous Plot

```python
##Correlation with Heatmap

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
##Joint Plot

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
##Pair Plot

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

```python
##Dist Plot

sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

# Categorical Plot

```python
##Count Plot

sns.countplot('sex',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
##Bar Plot

sns.barplot(x='total_bill',y='day',data=df)
```

```python
sns.barplot('sex','total_bill',data=df)
```

```python
##Box Plot

sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot('day','total_bill',hue='sex',data=df)
```

```python
##Violin Plot

sns.violinplot('total_bill','day',data=df)
```

```python
sns.violinplot('sex','tip',data=df)
```
