```python
df=pd.read_csv("mercedesbenz.csv", usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
## How many labels each variable have?

for col in df.columns:
    print(col, ':', len(df[col].unique()), 'labels')
```

```python
## Lets examine how many columns we will obtain after one hot encoding these variables

pd.get_dummies(df,drop_first=True).shape
```

```python
## Find out top 10 most frequent categories for the variable X2

df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
## Lets make a list with most frequent categories of the variable.

top_10 = [x for x in df.X2.value_counts().head(10).index]
top_10
```

```python
## Now we make the 10 binary variables

for label in top_10:
    df[label]=np.where(df['X2']==label,1,0)
    
df[['X2']+top_10].head(40)
```

```python
## Get the whole set of dummy variables, for all categorical variables

def one_hot_top_x(df, variables, top_x_labels):
    ## Function to create most dummy variables for the most frequent labels
    ## We can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        df[variables+'_'+label]=np.where(df[variables]==label,1,0)
        
## Read the data again
df=pd.read_csv('mercedesbenz.csv', usecols=['X1','X2','X3','X4','X5','X6'])

## Encode X2 into top 10 most frequent categories
one_hot_top_x(df,'X2',top_10)
df.head()
```

```python
## Lets do it for X1
## Step 1 - Find the count of categories

df.X1.value_counts().head(10)
```

```python
## Step 2 - Make the list

top_5 = [x for x in df.X1.value_counts().sort_values(ascending=False).head(5).index]
top_5
```

```python
## Step 3 - Make the list binary

for label in top_5:
    df[label]=np.where(df['X1']==label,1,0)

df[['X1']+top_5].head(20)
```

```python
## Step 4 - Call the function

df=pd.read_csv('mercedesbenz.csv', usecols=['X1','X2','X3','X4','X5','X6'])

one_hot_top_x(df,'X1',top_5)
df.head()
```
