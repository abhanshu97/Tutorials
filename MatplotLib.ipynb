```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Graph')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'b*--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot Graph')
```

```python
##Creating Subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
np.pi
```

```python
x=np.arange(0, 5*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'r')
```

```python
x=np.arange(0, 5*np.pi,0.1)
y=np.sin(x)
y1=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y,'g')
plt.title('Sine Wave')


plt.subplot(2,1,2)
plt.plot(x,y1,'b')
plt.title('Cosine Wave')

```

```python
x = [2,6,8,10]
y = [11,9,16,9]

x1 = [3,9,11]
y1 = [6,15,7]

plt.bar(x,y,align='center')
plt.bar(x1,y1)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Graph')
```

```python
x=np.array([2,20,12,23,31,14,27,4,5,28,29])
plt.hist(x)
plt.title('Histogram')
```

```python
##Box Plot

k=[np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(k,vert=True,patch_artist=True)
plt.title('Box Plot')
plt.show()
```

```python
## Pie Chart

labels = 'Python', 'C', 'Java', 'PHP'
sizes = [200,70,145,120]
color = ['green','yellow','orange','blue']
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=color, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```
