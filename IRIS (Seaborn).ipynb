```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#Joint Plot

sns.jointplot('sepal_length','petal_length',data=df,kind='reg')
```

```python
sns.jointplot('sepal_length','sepal_width',data=df,kind='reg')
```

```python
sns.jointplot('sepal_length','petal_width',data=df,kind='reg')
```

```python
#Pair Plot

sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='species')
```

```python
#Dist Plot

sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['petal_length'],kde=False)
```

```python
#Count Plot

sns.countplot('species',data=df)
```

```python
df['species'].value_counts()
```

```python
#Bar Plot

sns.barplot('species','sepal_length',data=df)
```

```python
sns.barplot('species','petal_length',data=df)
```

```python
sns.barplot('species','sepal_width',data=df)
```

```python
sns.barplot('species','petal_width',data=df)
```

```python
#Box Plot

sns.boxplot('species','petal_length',data=df,palette='rainbow')
```

```python
#Violin Plot

sns.violinplot('species','sepal_width',data=df)
```
